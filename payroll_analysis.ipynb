```python
import pandas as pd
import numpy as np
```

```python
#Enter location
location = 'DSB'
```

```python
file = r'/your_location/file.xlsx'
df = pd.read_excel(file)
df.head()
```

```python
#Formats dataset for weekly presentation
weekly = df
w = weekly.loc[(weekly['Date'] > '2022-12-31')]
w = w[w['Location'] == location]
w.head()
```

```python
#Creates weekly Summary
w_pvt = pd.pivot_table(w,
                     index=['Week #'],
                     values='Avg Daily Payroll',
                     aggfunc=np.sum,
                     margins=False,
                     margins_name='Total')

#Undo indexes generated by Pivot
w_pvt.to_clipboard(index=False)
w_pvt
```

```python
w_pvt.plot()
```

```python
#Formats dataset for monthly presentation
monthly = df
m = monthly.loc[(monthly['Date'] > '2022-12-31')]
m = m[m['Location'] == location]
g = (m
 .groupby([pd.Grouper(key='Date', freq='M')])
 .agg(Labor_Cost = ('Avg Daily Payroll', 'sum'))
)
g.to_clipboard(index=False)
g.head()
```
