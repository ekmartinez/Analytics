```python
#load necesary libries
import pandas as pd
import numpy as np
```

```python
#Enter location
location = 'DSB'
```

```python
#Load Dataset
f = r'/your_location/file.xlsx'

df = pd.read_excel(f)
df['Date'] = pd.to_datetime(df['Date'])

df.head()
```

```python
#Format dataset for daily presentation
daily = df
d = daily[(daily['Date'] > '2023-4-9') & (daily['Date'] < '2023-4-17')]
d = d[d['Location'] == location]

# #For some reason the Pivot Table adds timestamp to date, the following line removes timedate/
d['date2'] = d['Date'].dt.date
d
```

```python
#Creates Daily Summary
d_pvt = pd.pivot_table(d,
                     index=['date2', 'week_day'],
                     columns='Location',
                     values='net_sales',
                     aggfunc=np.sum,
                     margins=False,
                     margins_name='Total')

#Undo indexes generated by Pivot
d_pvt.reset_index(inplace=True)
d_pvt.to_clipboard()
d_pvt
```

```python
#Formats dataset for weekly presentation
weekly = df
w = weekly.loc[(weekly['Date'] > '2022-12-31')]
w = w[w['Location'] == location]
w.head()
```

```python
#Creates weekly Summary
w_pvt = pd.pivot_table(w,
                     index=['week'],
                     values='net_sales',
                     aggfunc=np.sum,
                     margins=False,
                     margins_name='Total')

#Undo indexes generated by Pivot
w_pvt.reset_index(inplace=True)
w_pvt.to_clipboard()
w_pvt
```

```python
#Formats dataset for monthly presentation
monthly = df
m = monthly.loc[(monthly['Date'] > '2022-12-31')]
m = m[m['Location'] == location]
g = (m
 .groupby([pd.Grouper(key='Date', freq='M')])
 .agg(Net_Sales = ('net_sales', 'sum'))
)
g.reset_index(inplace=True)
g.to_clipboard()
g.head()
```

```python
#Hotter Days
hot = df
hot = hot[hot['Location'] == location]

hot.head()
```

```python
#Takes the entire history to present on which week days sales are produced.
hot_pvt = pd.pivot_table(hot,
                     index=['week_day'],
                     values='net_sales',
                     aggfunc=np.sum,
                     margins=False,
                     margins_name='Total')

hot_pvt.to_clipboard()
hot_pvt
```

```python
hist = df
hist = hist[hist['Location'] == location]
hist = hist.loc[(hist['Date'] < '2023-4-1')]
hist.head()
```

```python


g_hist = (hist
 .groupby([pd.Grouper(key='Date', freq='M')])
 .agg(Net_Sales = ('net_sales', 'sum'))
)
g_hist.to_clipboard()
g_hist
```
